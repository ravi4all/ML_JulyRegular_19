```python
votes = [355,120,45,32]
parties = ['BJP','CONG','SP+BSP','Others']
```

```python
plt.figure(figsize=(10,6))
plt.pie(votes,labels=parties,counterclock=False,startangle=90,explode=(0,0,0.2,0),autopct='%1.2f%%')
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn")
plt.bar(parties,votes,color='red')
plt.title("Election Results 2019")
plt.xlabel("Parties")
plt.ylabel("Votes")
plt.show()
```
